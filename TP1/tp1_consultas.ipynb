## Consulta 1: Cuál es el estado que más descuentos tiene en total? y en promedio?


```python
import re

ordenes['billing_state'] = ordenes['billing_address'].str.extract(r',?\s*([A-Za-z]{2})\s*\d{5}', flags=re.IGNORECASE)[0].str.upper()

ordenes_filtradas = ordenes[ordenes['billing_state'].notnull()]

ordenes_filtradas = ordenes_filtradas[ordenes_filtradas['discount_amount'].notna()]

ordenes_filtradas = ordenes_filtradas[ordenes_filtradas['discount_amount'] > 0]

estados_cant_descuentos = ordenes_filtradas.groupby('billing_state')['discount_amount'].agg({'sum'})
estados_cant_descuentos.rename(columns={'sum': 'total_descuentos'}, inplace=True)
estado_mas_descuentos = estados_cant_descuentos.nlargest(1, columns='total_descuentos')

estados_prom_descuentos = ordenes_filtradas.groupby('billing_state')['discount_amount'].agg({'mean'})
estados_prom_descuentos.rename(columns={'mean': 'prom_descuentos'}, inplace=True)
estado_prom_descuentos = estados_prom_descuentos.nlargest(1, columns='prom_descuentos')

display(estado_mas_descuentos, estado_prom_descuentos)
```

## Consulta 2: ¿Cuáles son los 5 códigos postales más comunes para las órdenes con estado ‘Refunded’? ¿Y cuál es el nombre más frecuente entre los clientes de esas direcciones?


```python
ordenes['billing_postal_code'] = ordenes['billing_address'].str.extract(r'\s+(\d{5})\s*$')

filtro_codigos = ordenes['billing_postal_code'].notnull()
ordenes_filtradas_codigos = ordenes[filtro_codigos]

filtro_nan_estado = ordenes_filtradas_codigos['status'].notna()
ordenes_filtradas_codigos = ordenes_filtradas_codigos[filtro_nan_estado]

filtro_estado_undefined = ordenes_filtradas_codigos['status'].str.contains('undefined', case=False, na=False)
ordenes_filtradas_codigos = ordenes_filtradas_codigos[~filtro_estado_undefined]

ordenes_filtradas_codigos['status'] = ordenes_filtradas_codigos['status'].str.lower().str.strip()

filtro_estado_orden = ordenes_filtradas_codigos['status'] == 'refunded'
ordenes_filtradas_codigos = ordenes_filtradas_codigos[filtro_estado_orden]

codigos_cant_ordenes = ordenes_filtradas_codigos.groupby('billing_postal_code')['status'].agg({'count'})
codigos_cant_ordenes.rename(columns={'count': 'cant_ordenes'}, inplace=True)
codigos_mas_comunes = codigos_cant_ordenes.nlargest(5, columns='cant_ordenes')

ordenes_buscadas = codigos_mas_comunes.merge(ordenes_filtradas_codigos, on='billing_postal_code', how='inner')
clientes_buscados = clientes.merge(ordenes_buscadas, on='customer_id', how='inner')

filtro_nombre_nan = clientes_buscados['first_name'].notna()
clientes_buscados = clientes_buscados[filtro_nombre_nan]

filtro_nombre_undefined = clientes_buscados['first_name'].str.contains('undefined', case=False, na=False)
clientes_buscados = clientes_buscados[~filtro_nombre_undefined]

filtro_id_nan = clientes_buscados['customer_id'].notna()
clientes_buscados = clientes_buscados[filtro_id_nan]

clientes_buscados['first_name'] = clientes_buscados['first_name'].str.lower().str.strip()

nombres_cant_apariciones = clientes_buscados.groupby('first_name')['customer_id'].agg({'count'})
nombres_cant_apariciones.rename(columns={'count': 'cant_apariciones'}, inplace=True)
nombre_mas_apariciones = nombres_cant_apariciones.nlargest(1, columns='cant_apariciones')

display(codigos_mas_comunes, nombre_mas_apariciones)
```

## Consulta 3: Para cada tipo de pago y segmento de cliente, devolver la suma y el promedio expresado como porcentaje, de clientes activos y de consentimiento de marketing.


```python
filtro_id_nan_ordenes = ordenes['customer_id'].notna()
ordenes_filtradas = ordenes[filtro_id_nan_ordenes]

filtro_pago_nan = ordenes_filtradas['payment_method'].notna()
ordenes_filtradas = ordenes_filtradas[filtro_pago_nan]

filtro_pago_undefined = ordenes_filtradas['payment_method'].str.contains('undefined', case=False, na=False)
ordenes_filtradas = ordenes_filtradas[~filtro_pago_undefined]

filtro_id_nan_clientes = clientes['customer_id'].notna()
clientes_filtrados = clientes[filtro_id_nan_clientes]

filtro_segmento_nan = clientes_filtrados['customer_segment'].notna()
clientes_filtrados = clientes_filtrados[filtro_segmento_nan]

filtro_segmento_undefined = clientes_filtrados['customer_segment'].str.contains('undefined', case=False, na=False)
clientes_filtrados = clientes_filtrados[~filtro_segmento_undefined]

filtro_activo_nan = clientes_filtrados['is_active'].notna()
clientes_filtrados = clientes_filtrados[filtro_activo_nan]

filtro_marketing_nan = clientes_filtrados['marketing_consent'].notna()
clientes_filtrados = clientes_filtrados[filtro_marketing_nan]

ordenes_clientes = ordenes_filtradas.merge(clientes_filtrados, on='customer_id', how='inner')

ordenes_clientes['payment_method'] = ordenes_clientes['payment_method'].str.lower().str.strip()
ordenes_clientes['customer_segment'] = ordenes_clientes['customer_segment'].str.lower().str.strip()

clientes_unicos_por_grupo = ordenes_clientes[['customer_id', 'payment_method', 'customer_segment', 'is_active', 'marketing_consent']].drop_duplicates(subset=['customer_id', 'payment_method', 'customer_segment'])

ordenes_clientes_agrupados_activos = clientes_unicos_por_grupo.groupby(['payment_method', 'customer_segment'])['is_active'].agg(['sum', 'mean'])
ordenes_clientes_agrupados_marketing = clientes_unicos_por_grupo.groupby(['payment_method', 'customer_segment'])['marketing_consent'].agg(['sum', 'mean'])

ordenes_clientes_agrupados_activos.rename(columns={'sum': 'total_clientes_activos', 'mean': 'porcentaje_clientes_activos'}, inplace=True)
ordenes_clientes_agrupados_marketing.rename(columns={'sum': 'total_clientes_marketing', 'mean': 'porcentaje_clientes_marketing'}, inplace=True)

ordenes_clientes_agrupados_activos = ordenes_clientes_agrupados_activos.rename_axis(['tipo_pago', 'segmento_cliente'])
ordenes_clientes_agrupados_marketing = ordenes_clientes_agrupados_marketing.rename_axis(['tipo_pago', 'segmento_cliente'])

ordenes_clientes_agrupados_activos['porcentaje_clientes_activos'] = ordenes_clientes_agrupados_activos['porcentaje_clientes_activos'] * 100
ordenes_clientes_agrupados_marketing['porcentaje_clientes_marketing'] = ordenes_clientes_agrupados_marketing['porcentaje_clientes_marketing'] * 100

ordenes_clientes_agrupados_activos['total_clientes_marketing'] = ordenes_clientes_agrupados_marketing['total_clientes_marketing']
ordenes_clientes_agrupados_activos['porcentaje_clientes_marketing'] = ordenes_clientes_agrupados_marketing['porcentaje_clientes_marketing']

resultado = ordenes_clientes_agrupados_activos

resultado
```

## Consulta 4: Para los productos que contienen en su descripción la palabra “stuff” (sin importar mayúsculas o minúsculas), calcular el peso total de su inventario agrupado por marca, mostrar sólo la marca y el peso total de las 5 más pesadas.


```python
filtro_descripcion_stuff = productos['description'].str.contains('stuff', case=False, na=False)
productos_con_stuff = productos[filtro_descripcion_stuff]

filtro_peso_nan = productos_con_stuff['weight_kg'].notna()
productos_con_stuff = productos_con_stuff[filtro_peso_nan]

filtro_marca_undefined = productos_con_stuff['brand'].str.contains('undefined', case=False, na=False)
productos_con_stuff = productos_con_stuff[~filtro_marca_undefined]

filtro_marca_nan = productos_con_stuff['brand'].notna()
productos_con_stuff = productos_con_stuff[filtro_marca_nan]

productos_por_marca = productos_con_stuff.groupby('brand')['weight_kg'].agg({'sum'})
productos_por_marca.rename(columns={'sum': 'total_peso'}, inplace=True)
marcas_mas_pesadas = productos_por_marca.nlargest(5, columns='total_peso')

marcas_mas_pesadas
```

## Consulta 5: Para cada categoría padre devolver aquella subcategoría que se haya creado hace más tiempo, indicando el nombre y la fecha de creación


```python
filtro_subcategoria_nan = categorias['category_name'].notna()
categorias_filtradas = categorias[filtro_subcategoria_nan]

filtro_subcategoria_undefined = categorias_filtradas['category_name'].str.contains('undefined', case=False, na=False)
categorias_filtradas = categorias_filtradas[~filtro_subcategoria_undefined]

filtro_categoria_nan = categorias_filtradas['parent_category'].notna()
categorias_filtradas = categorias_filtradas[filtro_categoria_nan]

filtro_categoria_undefined = categorias_filtradas['parent_category'].str.contains('undefined', case=False, na=False)
categorias_filtradas = categorias_filtradas[~filtro_categoria_undefined]

filtro_creacion_nan = categorias_filtradas['created_at'].notna()
categorias_filtradas = categorias_filtradas[filtro_creacion_nan]

filtro_creacion_undefined = categorias_filtradas['created_at'].str.contains('undefined', case=False, na=False)
categorias_filtradas = categorias_filtradas[~filtro_creacion_undefined]

categorias_filtradas['created_at'] = pd.to_datetime(categorias_filtradas['created_at'], format='mixed')

categorias_filtradas['parent_category'] = categorias_filtradas['parent_category'].str.lower().str.strip()

categorias_mas_antiguas = categorias_filtradas.loc[categorias_filtradas.groupby('parent_category')['created_at'].idxmin()]

resultado = categorias_mas_antiguas[['parent_category', 'category_name', 'created_at']]

resultado
```

## Consulta 6: Para cada tipo de movimiento de los registros de inventario, devolver la razón más frecuente de ese movimiento


```python
filtro_movimiento_nan = registros_inventario['movement_type'].notna()
registros_filtrados = registros_inventario[filtro_movimiento_nan]

filtro_movimiento_undefined = registros_filtrados['movement_type'].str.contains('undefined', case=False, na=False)
registros_filtrados = registros_filtrados[~filtro_movimiento_undefined]

registros_filtrados['movement_type'] = registros_filtrados['movement_type'].str.lower().str.strip()

filtro_razon_nan = registros_filtrados['reason'].notna()
registros_filtrados = registros_filtrados[filtro_razon_nan]

filtro_razon_undefined = registros_filtrados['reason'].str.contains('undefined', case=False, na=False)
registros_filtrados = registros_filtrados[~filtro_razon_undefined]

registros_filtrados['reason'] = registros_filtrados['reason'].str.lower().str.strip()

razon_por_movimiento_conteo = registros_filtrados.groupby(['movement_type', 'reason']).size().reset_index(name='cant_registros')

razon_mas_frecuente_por_movimiento = razon_por_movimiento_conteo.loc[razon_por_movimiento_conteo.groupby('movement_type')['cant_registros'].idxmax()]

razon_mas_frecuente_por_movimiento
```

## Consulta 7: Cuales son los 3 productos que más descuentos tienen en total? Y entre ellos, cuál es el de mayor precio (unitario)?


```python
filtro_descuento_nan = orden_productos['discount_amount'].notna()
orden_productos_filtrados = orden_productos[filtro_descuento_nan]

filtro_precio_nan = orden_productos_filtrados['unit_price'].notna()
orden_productos_filtrados = orden_productos_filtrados[filtro_precio_nan]

filtro_id_nan = orden_productos_filtrados['product_id'].notna()
orden_productos_filtrados = orden_productos_filtrados[filtro_id_nan]

orden_productos_agrupados = orden_productos_filtrados.groupby('product_id')['discount_amount'].agg({'sum'})
orden_productos_agrupados.rename(columns={'sum': 'total_descuentos'}, inplace=True)
productos_mas_descuentos = orden_productos_agrupados.nlargest(3, columns='total_descuentos')

productos_info = productos.merge(productos_mas_descuentos, left_on='product_id', right_index=True, how='inner')

productos_info[['product_id', 'product_name', 'price', 'total_descuentos']]

producto_mayor_precio_unitario = productos_info.loc[productos_info['price'].idxmax()]

producto_mayor_precio_unitario[['product_id', 'product_name', 'price', 'total_descuentos']]

display(productos_info, producto_mayor_precio_unitario)
```

## Consulta 8: Cuales son los 3 productos con peor rating de opiniones? Y en promedio? (considerando como malo un rating de menos de 3 puntos, y considerando como válida una opinión que tenga un comentario que no sea vacío)

```python
filtro_rating_nan = opiniones['rating'].notna()
opiniones_filtradas = opiniones[filtro_rating_nan]

filtro_comentario_nan = opiniones_filtradas['comment'].notna()
opiniones_filtradas = opiniones_filtradas[filtro_comentario_nan]

filtro_comentario_undefined = opiniones_filtradas['comment'].str.contains('undefined', case=False, na=False)
opiniones_filtradas = opiniones_filtradas[~filtro_comentario_undefined]

filtro_rating_bajos = opiniones_filtradas['rating'] < 3
opiniones_filtradas = opiniones_filtradas[filtro_rating_bajos]

opiniones_por_producto = opiniones_filtradas.groupby('product_id')['rating'].agg({'sum'})
opiniones_por_producto.rename(columns={'sum': 'rating_total'}, inplace=True)
productos_peor_puntados = opiniones_por_producto.nlargest(3, columns='rating_total')

productos_info = productos.merge(productos_peor_puntados, left_on='product_id', right_index=True, how='inner')

productos_info[['product_id', 'product_name', 'rating_total']]

opiniones_por_producto_promedio = opiniones_filtradas.groupby('product_id')['rating'].agg({'mean'})
opiniones_por_producto_promedio.rename(columns={'mean': 'rating_promedio'}, inplace=True)
productos_peor_puntuados_promedio = opiniones_por_producto_promedio.nlargest(3, columns='rating_promedio')

productos_promedio_info = productos.merge(productos_peor_puntuados_promedio, left_on='product_id', right_index=True, how='inner')

productos_promedio_info[['product_id', 'product_name', 'rating_promedio']]

display(productos_info, productos_promedio_info)
```
